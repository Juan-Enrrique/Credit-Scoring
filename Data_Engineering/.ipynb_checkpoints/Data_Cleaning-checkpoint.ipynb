```python
def dicc_column(df_column, dicc_value = None, sep = ',', replace = ''):
    if dicc_value is None:
        dicc_value = set()
    column = df_column.dropna()
    for i in column:
        if sep not in i and i not in dicc_value:
            dicc_value.add(i)
        else:
            clean_data = i.replace(replace, '').split(sep)
            dicc_value.update(map(str.strip, clean_data))
    return dict(enumerate(sorted(dicc_value)))
```

##### Object Column - NaN values

```python
def Reassign(df, groupby, column, inplace = True):
    def fill_mode(df, groupby, column, inplace = True):
        #Convert None to NaN
        if df[column].isin([None]).sum():
            df[column] = df[column].replace(None, np.NaN)

        #Using pd.Series.mode() instead of stats.mode
        result = df.groupby(groupby)[column].transform(lambda x: x.fillna(x.mode()[0] if not x.mode().empty else np.NaN))
        if inplace:
            df[column] = result
        else:
            return result
    if inplace:
        #Msg before assign values
        if df[column].isna().sum():
            nan_count = df[column].isna().sum()
            print(f'\nBefore Assigning: {column}:', f'have {nan_count} NaN Values', end = '\n')
        #Print before assign
        grouped_values = df.groupby(groupby)[column].apply(list)
        print(f'\nBefore Assigning {column}:\n', *grouped_values.head().values, sep = '\n', end = '\n')

        fill_mode(df, groupby, column, inplace)

        #Msg After assign values
        if df[column].isna().sum():
            nan_count = df[column].isna().sum()
            print(f'\nAfter Assigning: {column}:', f'have {nan_count} NaN Values', end = '\n')

        # Print after assign
        grouped_values = df.groupby(groupby)[column].apply(list)
        print(f'\nAfter Assigning {column}:\n', *grouped_values.head(), sep = '\n', end = '\n')
    else:
        return fill_mode(df, groupby, column, inplace)
```

```python
def iqr(df, column, multiply = 1.5):
    #Calculate Q1 y Q3 una sola vez
    q1, q3 = df[column].quantile([0.25, 0.75])
    iqr = q3 - q1

    #Calculate lower and upper limits
    lower = q1 - iqr * multiply
    upper = q3 + iqr * multiply

    #Use 'between' to improve filtering
    outliers = (~df[column].between(lower, upper)).sum()

    print('Outliers: ', outliers)
    return lower, upper
```

```python
def ReassignNum(df, groupby, column, inplace=True):      
    import scipy.stats as stats

    # Identify Wrong values Range
    def group_min_max(df, groupby, column):            
        cur = df[df[column].notna()].groupby(groupby)[column].apply(list)
        modes = cur.apply(lambda x: stats.mode(x, keepdims=True).mode[0])
        return modes.min(), modes.max()
    
    # Assigning Wrong values
    def to_nan_fill_mode(df, groupby, column, inplace=True):
        df_dropped = df[df[column].notna()].groupby(groupby)[column].apply(list)
        modes = df_dropped.apply(lambda x: stats.mode(x, keepdims=True).mode[0])
        mini, maxi = modes.min(), modes.max()

        # Assign Wrong Values to NaN
        col = df[column].apply(lambda x: np.NaN if ((x < mini) | (x > maxi)) else x)

        # Fill with local mode
        mode_group = df.groupby(groupby)[column].transform(lambda x: x.mode()[0] if not x.mode().empty else np.NaN)
        result = col.fillna(mode_group)

        # Inplace or return
        if inplace:
            df[column] = result
        else:
            return result
    
    # Run      
    if inplace:   
        # Before Assigning NaN values   
        if df[column].value_counts(dropna=False).index.isna().sum():
            x = df[column].value_counts(dropna=False).loc[[np.NaN]]
            print(f'\nBefore Assigning: {column}:', f'have {x.values[0]} NaN Values', end='\n')
            
        print("\nExisting Min, Max Values:", df[column].apply([min, max]), sep='\n', end='\n')       
        mini, maxi = group_min_max(df, groupby, column)        
        print(f"\nGroupby by {groupby}'s Actual min, max Values:", f'min:\t{mini},\nmax:\t{maxi}', sep='\n', end='\n')        
        
        a = df.groupby(groupby)[column].apply(list) 
        print(f'\nBefore Assigning {column}:\n', *a.head().values, sep='\n', end='\n')
        
        # Assigning
        to_nan_fill_mode(df, groupby, column, inplace)
        
        # After Assigning NaN values
        if df[column].value_counts(dropna=False).index.isna().sum():
            y = df[column].value_counts(dropna=False).loc[[np.NaN]]
            print(f'\nAfter Assigning: {column}:', f'have {y.values[0]} NaN Values', end='\n')
        
        b = df.groupby(groupby)[column].apply(list)
        print(f'\nAfter Assigning {column}:\n', *b.head().values, sep='\n', end='\n')
    else:   
        # Show
        return to_nan_fill_mode(df, groupby, column, inplace)
```
